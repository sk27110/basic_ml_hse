### Задание 12

Выполните задание 11 с помощью циклов. Сравните время выполнения

```python
# ваш код здесь
import time
start = time.time()

array = df['race/ethnicity'].unique()
num_points = {x:0 for x in array}
cnt = {x:0 for x in array}

for index, row in df.iterrows():
  cnt[row['race/ethnicity']] +=1
  num_points[row['race/ethnicity']]+=row['reading score']
for i in array:
  print(i, num_points[i]/cnt[i])

end = time.time() - start
print("time:", end)
```

### Задание 14

Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице `df` под названием `Math_PassStatus` и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
# ваш код здесь
df['Math_PassStatus'] = df['math score'].apply(lambda x: 'P' if x >= passmark else 'F')
print(df['Math_PassStatus'].value_counts())

df['Reading_PassStatus'] = df['reading score'].apply(lambda x: 'P' if x >= passmark else 'F')
print(df['Reading_PassStatus'].value_counts())

df['Writing_PassStatus'] = df['writing score'].apply(lambda x: 'P' if x >= passmark else 'F')
print(df['Writing_PassStatus'].value_counts())
```

### Задание 16

Переведем баллы в оценки

**Система перевода баллов в оценки**
*   [90;100] = A
*   [80;90) = B
*   [70;80) = C
*   [60;70) = D
*   [50;60) = E
*   [0;50) = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец `Grade` и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

```python
import numpy

def GetGrade(average_mark):
    if average_mark >=90 and average_mark <=100:
      return 'A'
    elif average_mark >=80 and average_mark <90:
      return 'B'
    elif average_mark >=70 and average_mark <80:
      return 'C'
    elif average_mark >=60 and average_mark <70:
      return 'D'
    elif average_mark >=50 and average_mark <60:
      return 'E'
    else:
      return 'F'

# ваш код здесь
df['Grade'] = df[['math score', 'reading score', 'writing score']].mean(axis=1).apply(GetGrade)
df
```
